**Findings and Issues**

Findings and Issues

1.   R-squared is 0.7644%, meaning that the model describes 76.44% of the variation in the dependent variable; this is rather high (but not as high as in the G7 model).
2. However, there are reasons to doubt the accuracy of the weights (cofficients). This is because of the other R2 values. While our R-squared (within and overall) is very high, there is an alarmingly low R2 value in R2 (between). This means that the model can only explain data well if it contains one country at a time (intra-household explainability is better than its inter-household explainability). R-squared (between) has a value of 0.1383; this R-Squared shows that only 13.83% of the variation in overall readiness is explained if the model contains more than one country.
3.  The OLS Model is statistically significant at the 95% confidence level given that p = 0.0000.
4. None of the variables are statistically insignificant. However, if executing the model with **all** variables at once is attempted, there is an "Absorption Error" and Python states that it cannot perform linear regression due to high multicollinearity of variables. This means that there may be several irrelevant and confounding features in the whole dataset.

## Hyperparameter Training: RFE Feature Selection

```python
#Models to be refined through RFE
models = {"Linear Regression": LinearRegression(),
          "SVM": SVR(kernel="linear"),
          "Random Forest": RandomForestRegressor(random_state=random_state)
          }

# define parameters
params = {"Linear Regression" : {},
          "SVM": {"gamma": ["scale", "auto"],
                  "C": [0.01, 1, 10],
                  "epsilon": np.arange(0.1, 1.1, 0.1)},
          "Random Forest": {"n_estimators": np.arange(100, 600, 100)}
          }
```

## Model Training: Identifying the Best Model through RMSE

```python
def optim_model(model_key, feature_selection_method=None,
                X_train=X_train, y_train=y_train,
                X_val=X_val, y_val=y_val,
                X_test=X_test, y_test=y_test):
    print(f"Optimizing: {model_key}")
    
    # select features on train set
    if feature_selection_method is "RFECV":
        rfecv = RFECV(models[model_key], step=1, 
                      scoring="neg_root_mean_squared_error", 
                      cv=5, verbose=0, n_jobs=-1)
        rfecv.fit(X_train, y_train)

        n_features= rfecv.n_features_ 
        feature_names = rfecv.get_feature_names_out()
        
    elif feature_selection_method is None: 
        n_features=len(X_val.columns)
        feature_names = X_val.columns 

    # gridsearch selected features
    # tune hyperparameters on validation set
    cv_rfe = GridSearchCV(models[model_key], 
                          param_grid=params[model_key], 
                          cv=5, 
                          scoring="neg_root_mean_squared_error", 
                          verbose=0,
                          n_jobs=-1)
    X_rfe = X_val[feature_names]
    cv_rfe.fit(X_rfe, y_val)
    top_model_rmse1 = cv_rfe.best_estimator_
        
        
    print(f"The optimal number of features are {n_features}")
    print(f"These features are: {feature_names}")
    print("\n")
    
    # return best RMSE scores:
    h_test = top_model_rmse1.predict(X_test[feature_names])
    
    test_score = np.sqrt(mean_squared_error(y_test, h_test))
    
    results_dict = {"Model": top_model_rmse1, 
                    "Feature Names": feature_names, 
                    "Test Score": test_score}  
      
    return results_dict
```

```python
# automate training process
def train(model_keys, fs_params):
    # init containers
    results = collections.defaultdict()
    optimal_score = np.inf
    
    # train
    for model, params in zip(model_keys, fs_params):
        temp_result = optim_model(model, params)
        results.setdefault(model, []).append(temp_result["Test Score"])
        if temp_result["Test Score"] < optimal_score:
            optimal_score = temp_result["Test Score"]
            optimal_feature_names = temp_result["Feature Names"]
            optimal_model = temp_result["Model"]
    # append panel regression test score
    results.setdefault("Fixed Effects", []).append(panelreg_test_score_rmse)
    
    # sort results by test score
    sorted_results = {model_name: score_val for model_name, score_val in 
                      (sorted(results.items(), key=lambda item:item[1]))}
    
    return optimal_model, optimal_feature_names, sorted_results
```

```python
top_model_rmse1, top_feature_names_rmse1, results = train(list(models.keys()), 
                                              ["RFECV", "RFECV", "RFECV"])
```

```python
# training process
def train(model_keys, fs_params):
    # init containers
    results = collections.defaultdict()
    optimal_score = np.inf
    
    # train
    for model, params in zip(model_keys, fs_params):
        temp_result = optim_model(model, params)
        results.setdefault(model, []).append(temp_result["Test Score"])
        if temp_result["Test Score"] < optimal_score:
            optimal_score = temp_result["Test Score"]
            optimal_feature_names = temp_result["Feature Names"]
            optimal_model = temp_result["Model"]
    # append panel regression test score
    results.setdefault("Fixed Effects", []).append(panelreg_test_score_mae)
    
    # sort results by test score
    sorted_results = {model_name: score_val for model_name, score_val in 
                      (sorted(results.items(), key=lambda item:item[1]))}
    
    return optimal_model, optimal_feature_names, sorted_results
```

# Final Model

Even though the Random Forest Regression model from the odd-and-even-year-based split has better RMSE, R2, and MAE scores, we still prefer the sequential-based split. This is because the data is linear and avoiding the years in sequence may inhibit the ability of the model to gauge linear progression over the years.

Incidentally, the sequential-based  Random Forest Regression model does not use the same characteristics across. We must thus select the best features to use in the model based on the R2, RMSE, and MAE scores. The RFR model with the best performing RMSE uses 6 characteristics, while the RFR model with the best performing MAE and R2 used 5 characteristics. 

We use the RFR model with 6 characteristics and the best RMSE. This is because of two reasons:
1. RMSE penalizes the model more if the data has issues of bias, variance, and outliers. Since the ASEAN region contains both very rich and very poor countries, we want to limit the degree of influence that outliers have such that we can have a better model. As stated by [JJ (2016)](https://medium.com/human-in-a-machine-world/mae-and-rmse-which-metric-is-better-e60ac3bde13d) and in our lessons, RMSE is more critical than MAE as it places a larger weight to errors.
2. When looking at model explainability (as shown in the code below) our RFR model with 6 characteristics still has quite a high explainability at an R2 of 99%, meaning that we are not trading off much by choosing RMSE over R2 as a metric of interest (but may trade off more by choosing R2 over RMSE given the heterogeneity of ASEAN countries).
3. A higher number of variables is kept by RFR, which indicates that it can see the effects of other variables more.

### Interpretation of Global Explainability
The most important feature for ASEAN countries is exposure. Exposure is the extent to which stress is imposed to human society by climate change. These are the external physical factors that contribute to the vulnerability of a country to climate change. ASEAN countries are often referred to as one of the most vulnerable regions, given that most countries have long coastlines and large populations in low-lying areas, such as the Philippines, which regularly experience destructive typhoons (McElwee, 2021). Among the ASEAN countries, it is observed that the more moderate your exposure, the more climate change ready you are as a nation. However, if the country experiences extremes, they lower the climate change readiness. 
  - It can be seen that if the value for exposure is extremely low, then it drives the propensity of overall readiness to be very low. This means that if a country has low risks of climate vulnerability, then these countries will invest less on climate change actions. In  a survey conducted by the Climate Change in Southeast Asia Programme (2021), majority of the countries have responded that their respective governments do not allocate sufficient resources to address climate change. 
  - In another study by the United Nations, it was found that ASEAN countries give most priority to reducing poverty and inequality, education, and economic growth to create more jobs. The lack of priority to climate change has lowered the overall readiness of the country, meaning that these countries do not get to invest in climate change actions. 
  - We can also see that despite some ASEAN countries having extremely high exposure to climate change, it is still expected that overall readiness will be low. This is also due to the fact that investment priority is not given to the challenge of climate change. 
  - Some possibilities as to why moderately low and moderately high exposure drive the propensity of overall readiness to be high is due to climate financing and funding initiatives such as the Green Climate Fund and the Adaptation Fund (AF), which distribute billions of dollars of pledged assistance for developing countries. ASEAN countries such as Cambodia, Indonesia, and Myanmar have received the highest total financing from the Green Climate fund.  

The second most important feature is water vulnerability. Water vulnerability is the degree to which access to water is susceptible to disruption due to climate change. About 300 million in the ASEAN region still have no access to safely managed or basic water services such as drinking water. Many countries in this region have also been experiencing increasing challenges of water pollution. This may be the reason why water vulnerability is the second most important feature. 
  - If access to clean water is unavailable, they are likely not to be ready to take on the effects of climate change. With extreme weather conditions that disrupt water availability, it does not come to a surprise that studies have found severe health consequences follow. The lack of water and sanitation is one of the leading causes of death among children below 5 years old, followed by diarrhea, which is also linked to drinking contaminated water and hygiene (UNICEF, 2022).


Economic readiness shows an association with climate change readiness. Consistent with the hypothesis, if a country, such that of the ASEAN, where generally serious economic and social issues prevail, would show low climate change readiness due to adequate resources directed towards addressing climate change and its risks. 
  - ASEAN countries experience challenges on rising commodity prices, supply bottlenecks, and vulnerabilities in financial sectors. These countries have yet to prioritize economic investments on sustainability such as green subsidies which is represented by the economic readiness feature. This may be the reason as to why lower economic readiness would drive the propensity of overall readiness to be low as well.
