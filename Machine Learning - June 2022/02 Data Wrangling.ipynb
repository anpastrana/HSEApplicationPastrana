## **02 Data Wrangling**

In this notebook we merged together the various datasets produced in the Pre-Wrangling phase.

```python
import pandas as pd
import numpy as np
```

```python
#Datasets
agri = pd.read_csv("https://raw.githubusercontent.com/idc0097/data103-project/main/data/mergedagriforfish.csv")
gdp = pd.read_csv("https://raw.githubusercontent.com/idc0097/data103-project/main/data/mergedgdp.csv")
publicvoice = pd.read_csv("https://raw.githubusercontent.com/idc0097/data103-project/main/data/mergedpublicvoiceandacc.csv")
readiness = pd.read_csv("https://raw.githubusercontent.com/idc0097/data103-project/main/data/mergedreadiness.csv")
vulnerability = pd.read_csv("https://raw.githubusercontent.com/idc0097/data103-project/main/data/mergedvulnerability.csv")
```

```python
#Recall that since Public Voice and Accountability does not have data for the years 1995, 1997, 1999, and 2001. This is why there are 68 null values; 17 * 4 = 68.
#This will be resolved in the imputing section of the project.

#Readiness and Vulnerability, with data from the years 1995 to 2019, have 425 rows. (17 countries * 25 years = 24 rows)
#GDP, Agri, and Public voice have data rows from the years 1995 to 2019, and have 442 rows. (17 countries * 26 years = 442 rows)
```

```python
#We first merge readiness and vulnerability together as they came from the same source. The product should be a dataset with 425 rows.
tableA = pd.merge(readiness, vulnerability)
display(tableA)
```

```python
tableA.isna().sum()
```

```python
#We next merge tableA with GDP, which has complete data. The ideal product should be something with 442 rows so we can smoothly merge the dataset with the rest.
tableA.dtypes
```

```python
gdp.dtypes
```

```python
tableB = pd.merge(tableA, gdp, how='right', on=['Country', 'Year'])
```

```python
display(tableB)
```

```python
#We should have 17 NaN values for 11 variables.
#This is because we have 11 variables for readiness and vulnerability for 17 countries which have no data for the year 2020. 
#This means that there should be 17 missing values for each readiness and vulnerability variable as there is no 2020 data.
tableB.isna().sum()
```

```python
#We have successfully merged GDP and tableA into tableB.
```

```python
#merging the rest
tableC = pd.merge(tableB, agri, how='right', on=['Country', 'Year'])
display(tableC)
```

```python
tableC.isna().sum()
#The null values are still the same from table B.
```

```python
tableD = pd.merge(tableC, publicvoice, how='right', on=['Country', 'Year'])
display(tableD)
```

```python
#There should be 17 + 68 = 85 null values in the publicvoiceandaccountabilityindex variable, and 17 null values in the readiness and vulnerability variables if we
#merged successfully.
tableD.isna().sum()
```

```python
#We have successfully merged our separate .CSVs together. 
```

```python
climatedata = tableD[['Country','Year','overallreadiness','socialreadiness','economicreadiness','governmentreadiness','ecosystemvulnerability','exposure','foodvulnerability',
'habitatvulnerability','healthvulnerability','watervulnerability','overallvulnerability','GDPPerCapita','PercGDPContributionOfAgriForestFish','publicvoiceandgovaccountabilityindex']]
display(climatedata)
```

```python
climatedata.isna().sum()
```

```python
#convert df to csv file
df = pd.DataFrame(climatedata, columns = ['Country', 'Year', 'overallreadiness','socialreadiness','economicreadiness','governmentreadiness','ecosystemvulnerability','exposure','foodvulnerability'
                                         ,'habitatvulnerability','healthvulnerability','watervulnerability','overallvulnerability','GDPPerCapita',
                                          'PercGDPContributionOfAgriForestFish','publicvoiceandgovaccountabilityindex'])
df.to_csv('climatedata.csv', encoding= 'utf-8',index=False)
```
