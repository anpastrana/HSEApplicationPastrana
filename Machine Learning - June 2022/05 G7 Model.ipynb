Findings and Issues (Simplified version, will be more detailed in final output)

1.   R-squared is 98.95%, meaning that the model describes 98.95% of the variation in the dependent variable; this is quite high.
2. However, there are reasons to doubt the accuracy of the weights (cofficients) below. This is because of the other R2 values. While our R-squared (within) is very high, this only means that the predictions of the changes in the dependent variable within countries is accurate. This means that the model can only explain data well if it contains one country at a time. R-squared (Between) has a negative value; this R-Squared shows how much of the variation in the dependent variable /between/ countries is explained by the model, and since it is low, it means that our model would perform very poorly in datasets containing several countries. The R-Squared (Overall) is the weighted average of the two, and it is negative, indicating a poor model. 
3.  The OLS Model is statistically significant at the 95% confidence level given that p = 0.0000.
4. We have the statistically insignificant feature of the Public Voice and Accountability Index at p = 0.2863. Its inclusion may have muddled the data.
5. If executing the model with **all** variables at once is attempted, there is an "Absorption Error" and Python states that it cannot perform linear regression due to high multicollinearity of variables. This means that there may be several irrelevant and confounding features in the whole dataset.

## Hyperparameter Training: RFE Feature Selection

```python
#Models to be refined through RFE
models = {"Linear Regression": LinearRegression(),
          "SVM": SVR(kernel="linear"),
          "Random Forest": RandomForestRegressor(random_state=random_state)
          }

# define parameters
params = {"Linear Regression" : {},
          "SVM": {"gamma": ["scale", "auto"],
                  "C": [0.01, 1, 10],
                  "epsilon": np.arange(0.1, 1.1, 0.1)},
          "Random Forest": {"n_estimators": np.arange(100, 600, 100)}
          }
```

## Model Training: Identifying the Best Model through RMSE

```python
def optim_model(model_key, feature_selection_method=None,
                X_train=X_train, y_train=y_train,
                X_val=X_val, y_val=y_val,
                X_test=X_test, y_test=y_test):
    print(f"Optimizing: {model_key}")
    
    # select features on train set
    if feature_selection_method is "RFECV":
        rfecv = RFECV(models[model_key], step=1, 
                      scoring="neg_root_mean_squared_error", 
                      cv=5, verbose=0, n_jobs=-1)
        rfecv.fit(X_train, y_train)

        n_features= rfecv.n_features_ 
        feature_names = rfecv.get_feature_names_out()
        
    elif feature_selection_method is None: 
        n_features=len(X_val.columns)
        feature_names = X_val.columns 

    # gridsearch selected features
    # tune hyperparameters on validation set
    cv_rfe = GridSearchCV(models[model_key], 
                          param_grid=params[model_key], 
                          cv=5, 
                          scoring="neg_root_mean_squared_error", 
                          verbose=0,
                          n_jobs=-1)
    X_rfe = X_val[feature_names]
    cv_rfe.fit(X_rfe, y_val)
    top_model_rmse1 = cv_rfe.best_estimator_
        
        
    print(f"The optimal number of features are {n_features}")
    print(f"These features are: {feature_names}")
    print("\n")
    
    # return best RMSE scores:
    h_test = top_model_rmse1.predict(X_test[feature_names])
    
    test_score = np.sqrt(mean_squared_error(y_test, h_test))
    
    results_dict = {"Model": top_model_rmse1, 
                    "Feature Names": feature_names, 
                    "Test Score": test_score}  
      
    return results_dict
```

```python
# automate training process
def train(model_keys, fs_params):
    # init containers
    results = collections.defaultdict()
    optimal_score = np.inf
    
    # train
    for model, params in zip(model_keys, fs_params):
        temp_result = optim_model(model, params)
        results.setdefault(model, []).append(temp_result["Test Score"])
        if temp_result["Test Score"] < optimal_score:
            optimal_score = temp_result["Test Score"]
            optimal_feature_names = temp_result["Feature Names"]
            optimal_model = temp_result["Model"]
    # append panel regression test score
    results.setdefault("Fixed Effects", []).append(panelreg_test_score_rmse)
    
    # sort results by test score
    sorted_results = {model_name: score_val for model_name, score_val in 
                      (sorted(results.items(), key=lambda item:item[1]))}
    
    return optimal_model, optimal_feature_names, sorted_results
```

```python
top_model_rmse1, top_feature_names_rmse1, results = train(list(models.keys()), 
                                              ["RFECV", "RFECV", "RFECV"])
```

```python
def optim_model(model_key, feature_selection_method=None,
                X_train=X_train, y_train=y_train,
                X_val=X_val, y_val=y_val,
                X_test=X_test, y_test=y_test):
    print(f"Optimizing: {model_key}")
    
    # select features on train set
    if feature_selection_method is "RFECV":
        rfecv = RFECV(models[model_key], step=1, 
                      scoring="neg_mean_absolute_error", 
                      cv=5, verbose=0, n_jobs=-1)
        rfecv.fit(X_train, y_train)

        n_features= rfecv.n_features_ 
        feature_names = rfecv.get_feature_names_out()
        
    elif feature_selection_method is None: 
        n_features=len(X_val.columns)
        feature_names = X_val.columns 

    # gridsearch selected features
    # tune hyperparameters on validation set
    cv_rfe = GridSearchCV(models[model_key], 
                          param_grid=params[model_key], 
                          cv=5, 
                          scoring="neg_mean_absolute_error", 
                          verbose=0,
                          n_jobs=-1)
    X_rfe = X_val[feature_names]
    cv_rfe.fit(X_rfe, y_val)
    top_model_mae2 = cv_rfe.best_estimator_
        
        
    print(f"The optimal number of features are {n_features}")
    print(f"These features are: {feature_names}")
    print("\n")
    
    # return best MAE scores:
    h_test = top_model_mae2.predict(X_test[feature_names])
    
    test_score = sklearn.metrics.mean_absolute_error(y_test, h_test)
    
    results_dict = {"Model": top_model_mae2, 
                    "Feature Names": feature_names, 
                    "Test Score": test_score}  
      
    return results_dict
```

```python
# training process
def train(model_keys, fs_params):
    # init containers
    results = collections.defaultdict()
    optimal_score = np.inf
    
    # train
    for model, params in zip(model_keys, fs_params):
        temp_result = optim_model(model, params)
        results.setdefault(model, []).append(temp_result["Test Score"])
        if temp_result["Test Score"] < optimal_score:
            optimal_score = temp_result["Test Score"]
            optimal_feature_names = temp_result["Feature Names"]
            optimal_model = temp_result["Model"]
    # append panel regression test score
    results.setdefault("Fixed Effects", []).append(panelreg_test_score_mae)
    
    # sort results by test score
    sorted_results = {model_name: score_val for model_name, score_val in 
                      (sorted(results.items(), key=lambda item:item[1]))}
    
    return optimal_model, optimal_feature_names, sorted_results
```

### Interpretation of Global Explainability

* Overall vulnerability is the most influential feature according to this graph of global explainability. The lower the values of of overall vulnerability to climate change, the higher the overall climate readiness of a G7 country. While extremely high overall vulnerability decreases the propensity of the overall readiness to be high. 
  - Despite having high overall vulnerability to extreme climate events, these countries have yet to expand their overall readiness to a level necessary for avoiding damages to the environment. Although stability in the rankings from the ND-Gain country index is consistent among the G7 countries, it has shown a similar trend in an assessment of wealthier countries, and it is that efforts to reduce climate vulnerability have plateaued (Klingerman, 2019). This is due, at least in part, to the power and ability of wealthier nations to pay for projects that are necessary to mitigate the effects of climate change. Another study has shown that wealthier countries can adapt better despite high vulnerability when they have higher levels of economic growth. 
  - G7 countries have the budget and can easily make effective use of investments for adaptation actions given their strong economy and efficient business environment. Even though these countries have extremely low vulnerability, they have the resources to fund for climate change actions, which increases the propensity of the overall readiness to be high.  

* The second-most influential feature is economic readiness. 
  - This is the variable indicating how much the economy has invested into being more sustainable (e.g. renewable energy sources, green subsidies). This variable also expounds on the investment climate that facilitates mobilizing capitals on sustainability. As G7 countries continue to lead in terms of economic development, we can also see that they invest more on sustainable actions. An example of this would be how Germany continues to expand the use of renewable energies and increase energy efficiency.
  - The lower the values of economic readiness, the lower the overall climate change readiness of a G7 country. However, even when this economic readiness has higher values (indicated in the red dots), it does not directly mean higher climate change readiness. 
 

* The third-most influential feature is social readiness. 
  - This is the variable indicating how much the culture and education system of the country treats climate change as a serious consideration. Social readiness is also the social conditions that aid society in making efficient use of investments to yield more benefits.
  - It is apparent that lower values of social readiness is accompanied by lower values of climate change readiness. Very high values have a positive impact on climate change readiness.


Note that for vulnerability indices, lower values mean *less vulnerable*, and higher values mean *more vulnerable.* 

* Let us look at the next three measures of vulnerability - health vulnerability (or the measure of how vulnerable a country's health sector is to climate change-aggravated illnesses like dengue and malaria), ecosystem vulnerability, habitat vulnerability, and food vulnerability (the vulnerability of a country's food supply to climate change).
  - All of these three features have more or less the same pattern in which higher levels of vulnerability drive the propensity of climate change readiness to a higher degree. This means that though the G7 countries experience great challenges, they are ready to respond and adopt solutions. 


* The higher the % of GDP contributed by Forestry, Agricultural, and Fishery sectors, the higher that country's overall climate change readiness.

* The lower the degree of exposure of a country to climate change, the higher its overall climate change readiness.
