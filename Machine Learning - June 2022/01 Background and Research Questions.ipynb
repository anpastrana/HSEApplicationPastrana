# **Introduction**

## **Importance of Studying Factors Affecting Climate Change Readiness**
We live in a time where there is an increasing emphasis on caring for the environment and protecting the world we call home. From climate change activists, Twitter trends, long Facebook posts, and news reports, the emphasis on climate change has been something that the youth these days consider a vital part of their lives. Some professors have even taught that care for the environment could decide whether a person purchases or patronizes a brand. However, outside social media and marketing, climate change influences human life more than some may expect. According to Adom & Amoani (2016), temperature increases may significantly affect economic and productivity growth, which may be aggregated by a nation's readiness to adapt. A report by the IPCC (2022) brings the reality of climate change closer to home by noting that, "Climate change impacts and risks are becoming increasingly complex and more difficult to manage. Multiple climate hazards will occur simultaneously, and multiple climatic and non-climatic risks will interact, resulting in compounding overall risk and risks cascading across sectors and regions. Some responses to climate change result in new impacts and risks." Additionally, Levin et al. (2022) mention that the same IPCC report emphasizes that some critical issues in the last few years include water insecurity, more enormous wildfires, and increasing temperatures. These issues have led to other problems, such as the spread of vector-borne and water-borne diseases.

## **Developed vs Developing Countries in terms of Climate Change Readiness**

It is without a question that developed countries such as the United States have high carbon emissions than developing countries since the Industrial revolution. Yet consequences are rarely experienced by these countries compared to developing countries, where resources and ability to risk management are vulnerable. With poor governance, poverty, food security, foreign debt, poor housing, and external environmental factors such as mining, bring out greater risks to developing countries. In the years 1991 and 2005, a significant portion of calamity-related deaths where from developing countries (Zorn, 2018). Furthermore, future adaptation to the increasing impact of weather-related natural disasters due to global climate change will also be more costly among developing countries.

## **Four Classes of Independent Variables Used in this Study**

### **Vulnerability**
  This paper focuses on four classes of independent variables: vulnerability, GDP Per Capita, Democracy, and % of GDP contributed by agriculture, forestry, and fishing industries. First vulnerability. Our data source defines this as the *"...propensity or predisposition of human societies to be negatively impacted by climate hazards."* The IPCC (2022) notes that "... a high proportion of species is vulnerable to climate change. Human and ecosystem vulnerability are interdependent...Current unsustainable development patterns are increasing exposure of ecosystems and people to climate hazards…" IPCC also notes that as global warming reaches + 1.5 degrees celsius, this will lead to increases in a multitude of hazards, which leads to increased risks for ecosystems and humans. However, "...the level of risk will depend on concurrent near- term trends in vulnerability, exposure, level of socioeconomic development and adaptation. (IPCC, 2022)"


### **GDP Per Capita**
Second, GDP per capita. According to the International IDEA (2021), "Climate shocks, such as natural disasters and extreme weather events, or problems such as food insecurity, can trigger occasional financial crises and economic decline, comparable with the experience under the Covid-19 pandemic." Both IDEA and research by Fuller (2020) agree that poorer countries will be more at risk due to the economic changes arising from climate change. Fuller (2020) in particular notes that, "low-income countries are more vulnerable to climate change than middle-and high-income countries even though they produce less greenhouse gases than the least vulnerable countries." On the contrary, research by Kompas et al. (2018) observed that if countries can comply with the terms of the Paris Climate Accord, they may save up to 17.489 billion dollars by the year 2100. According to Lo & Chow (2015), wealthier countries can prepare better for climate change because they prioritize this in their investments, making people feel safe due to the country's moves to prepare for the future.

### **Democracy**
Third, democracy. Specifically, the indicator of how good one's government is with public participation and accountability. Research by Povitkina (2018) suggests that "more democracy is only associated with lower CO2 emissions in low-corruption contexts. If corruption is high, democracies do not seem to do better than authoritarian regimes." Kamarck (2019) and Fiorino (2018) agree that a system that awards or punishes governments for their actions towards the environment must be in place. This ensures that these are not only self-seeking to political agendas but more so for the future and the future generations (Fiorno, 2018). Apart from the environmental effects, these systems may also be a way to help nations restore trust in their governments (Kamarck, 2019), which, based on research, is a critical factor that may influence a country's stance toward climate change and environmental action.

### **GDP Contributions from the Agriculture, Forestry, and Fishing Industries**
Fourth GDP but specifically towards agriculture, forestry, and fishing industries. Maulu et al. (2021) put this into context by noting that "...although climate change is a global food production risk, the associated risks on aquaculture are expected to differ across geographical or climatic zones, national economy, water environment, production systems, the scale of production, and cultured species of the aquaculture producers." On the other hand, Seino et al. (1998) note that global warming has three main effects on agriculture and forestry: increased CO2, other indirect effects, and rising sea levels threatening low-altitude areas. The FAO (2007) adds to this by noting that 60% of the Earth is covered by croplands, pastures and forests, all of which are threatened by climate change and global warming (e.g., flooding, drought). They note that ecosystems may be drastically affected in terms of habitats, spread of diseases, marine life populations, and even ocean circulation patterns, which may have an effect on a nation’s agriculture and production.
